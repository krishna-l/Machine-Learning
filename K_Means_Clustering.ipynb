**K-Means Clustering**

Creating fake income age clusters for N people in k clusters.

```
from numpy import random, array

def createClusters(N,k):
  random.seed(10) 
  pointsPerCluster = float(N)/k  #Total point per cluster
  X=[]
  for i in range(k):
    incomeCentroid = random.uniform(20000.0, 200000.0) #income centeroid
    ageCentroid = random.uniform(20.0, 70.0) #Age of mean 20 and std 70
    for j in range(int(pointsPerCluster)): # till the points per cluster we append a normal distribution of income and age to the list
      X.append([random.normal(incomeCentroid, 10000.0), random.normal(ageCentroid, 2.0)])
  X = array(X)
  return X
```

Applying K means for the data having 1000 people with 30 centroids. 

```
%matplotlib inline
from sklearn.cluster import KMeans # used to import K Means 
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale # used to scale the income and age to a value which is plotable
from numpy import float

data = createClusters(1000, 5) #create clusters of 1000 people with 30 clusters.

model = KMeans(n_clusters=5)

model = model.fit(scale(data)) # Normalizing the data by using scale.

print(model.labels_)

plt.figure(figsize=(8,6))
plt.scatter(data[:,0], data[:,1], c=model.labels_.astype(float)) #color is chosen from the labels in the model converted to float.
plt.show()
```
